```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

from src.rf import feature_selection_and_plot
```

```python
#LOAD DATASETS
train = pd.read_csv('./data/derived/feature_selection/training_data.csv', index_col=0)
test = pd.read_csv('./data/derived/feature_selection/test_data.csv', index_col=0)
combined = pd.concat([train, test])

#DEFINE RECENCY
recency_threshold = 24 * 30.44

# Label the data
combined['recency'] = np.where(combined['TSI_days'] <= recency_threshold, 1, 0)

#DEFINE X AND Y
X_train = combined.drop(columns=['recency', 'TSI_category', 'TSI_days', 'RENAME_ID'])
y_train = combined['recency']
```

```python
selected_features = feature_selection_and_plot(X_train, y_train, model_type= RandomForestClassifier, output_path='./outputs/features/feature_importance_binary.png')
```
