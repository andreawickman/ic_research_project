```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

from src.rf import feature_selection_and_plot
```

```python
#LOAD DATASETS
train = pd.read_csv('./data/derived/feature_selection/training_data.csv', index_col=0)
test = pd.read_csv('./data/derived/feature_selection/test_data.csv', index_col=0)
combined = pd.concat([train, test])

# Convert TSI_days to TSI_years
combined['TSI_years'] = combined['TSI_days'] / 365

# Transform TSI_years using the square root
combined['sqrt_TSI_years'] = np.sqrt(combined['TSI_years'])
```

```python
#DEFINE X AND Y
X_train = combined.drop(columns=['sqrt_TSI_years', 'TSI_years', 'TSI_category', 'TSI_days', 'RENAME_ID'])
y_train = combined['sqrt_TSI_years']
```

```python
selected_features = feature_selection_and_plot(X_train, y_train, model_type = RandomForestRegressor, output_path= './outputs/features/feature_importance_regressor.png')
```
